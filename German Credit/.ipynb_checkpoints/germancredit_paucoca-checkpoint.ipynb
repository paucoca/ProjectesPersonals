```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from ucimlrepo import fetch_ucirepo

from sklearn.model_selection import train_test_split
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier

from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, confusion_matrix, roc_curve
import warnings; warnings.simplefilter('ignore')


RANDOM_STATE = 42
sns.set_theme(style='whitegrid')
```

## 1. INTRODUCTION

Credit risk is the probability of not repaying a loan. We model it as binary classification:
- `default=1`: bad credit
- `default=0`: good credit

Economic motivation: information asymmetry and adverse selection.

Goal: compare Logistic Regression vs Random Forest vs XGBoost.

## 2. DATA DESCRIPTION

Note: The variable < 0 DM indicates that the applicant's checking account has a negative balance or is currently overdrawn in Deutsche Marks.

```python
statlog_german_credit_data = fetch_ucirepo(id=144)

X = statlog_german_credit_data.data.features.copy()
y_df = statlog_german_credit_data.data.targets.copy()

print(statlog_german_credit_data.variables)
```

```python
# Rename columns to descriptive names
rename_map = {
    'Attribute1': 'checking_status',
    'Attribute2': 'duration_months',
    'Attribute3': 'credit_history',
    'Attribute4': 'purpose',
    'Attribute5': 'credit_amount',
    'Attribute6': 'savings_account',
    'Attribute7': 'employment_since',
    'Attribute8': 'installment_rate',
    'Attribute9': 'personal_status_sex',
    'Attribute10': 'other_debtors',
    'Attribute11': 'residence_since',
    'Attribute12': 'property_type',
    'Attribute13': 'age_years',
    'Attribute14': 'other_installment_plans',
    'Attribute15': 'housing',
    'Attribute16': 'existing_credits_bank',
    'Attribute17': 'job_type',
    'Attribute18': 'liable_people',
    'Attribute19': 'telephone',
    'Attribute20': 'foreign_worker'
}
X = X.rename(columns=rename_map)

# y must be 1D (one label per row)
y = y_df['class'] if 'class' in y_df.columns else y_df.iloc[:, 0]

display(X.head())
display(y.head())
print('X shape:', X.shape)
print('y shape:', y.shape)
```

```python

# Convert target: class 1=good, 2=bad -> default 0=good, 1=bad
y_default = (pd.to_numeric(y, errors='coerce') == 2).astype(int)

class_table = pd.DataFrame({
    'count': y_default.value_counts().sort_index(),
    'percentage': (y_default.value_counts(normalize=True).sort_index() * 100).round(2)
})
display(class_table)

sns.countplot(x=y_default, palette='viridis')
plt.title('Class distribution: default')
plt.xlabel('default (0=good, 1=bad)')
plt.show()
```

**Interpretation:** The dataset is imbalanced (more good than bad), so we evaluate more than accuracy.

Accuracy is misleading because a model can ignore all "Bad" risks and still get a high score just by guessing "Good" for everyone.

We will also use: 

- Precision: Of all the borrowers your model predicted as "Bad," how many were actually bad?

- Recall: Of all the actual "Bad" borrowers in your data, how many did your model successfully catch?

## 3. DATA PREPROCESSING

To ensure a robust and reliable evaluation of the German credit model, the dataset was partitioned using train_test_split with a 30% test size, reserving the remaining 70% for model training. A random_state of 42 was implemented to guarantee that the split is reproducible for future testing, while the stratify=y_default parameter was applied to preserve the original distribution of "Good" and "Bad" credit risks across both subsets.

```python
categorical_cols = X.select_dtypes(include=['object', 'category']).columns.tolist()
numeric_cols = X.select_dtypes(exclude=['object', 'category']).columns.tolist()

X_train, X_test, y_train, y_test = train_test_split(
    X, y_default, test_size=0.30, random_state=RANDOM_STATE, stratify=y_default
)

print('Categorical columns:', categorical_cols)
print('Numeric columns:', numeric_cols)
print('Train/Test:', X_train.shape, X_test.shape)
```

## 4. EXPLORATORY DATA ANALYSIS

```python
display(X[numeric_cols].describe().T)
```

```python
eda_df = X.copy()
eda_df['default'] = y_default

fig, axes = plt.subplots(1, 3, figsize=(15, 4))
sns.boxplot(data=eda_df, x='default', y='duration_months', ax=axes[0], palette='Set2')
axes[0].set_title('Duration vs default')
sns.boxplot(data=eda_df, x='default', y='credit_amount', ax=axes[1], palette='Set2')
axes[1].set_title('Credit amount vs default')
sns.boxplot(data=eda_df, x='default', y='age_years', ax=axes[2], palette='Set2')
axes[2].set_title('Age vs default')
plt.tight_layout()
plt.show()
```

```python
checking_map = {'A11': '< 0 DM', 'A12': '0 <= x < 200 DM', 'A13': '>= 200 DM', 'A14': 'no checking'}
history_map = {'A30': 'all paid duly', 'A31': 'all paid at this bank', 'A32': 'paid till now', 'A33': 'delay in past', 'A34': 'critical account'}
savings_map = {'A61': '<100 DM', 'A62': '100-500 DM', 'A63': '500-1000 DM', 'A64': '>=1000 DM', 'A65': 'unknown/no savings'}
employment_map = {'A71': 'unemployed', 'A72': '<1 year', 'A73': '1-4 years', 'A74': '4-7 years', 'A75': '>=7 years'}

tab1 = eda_df.groupby('checking_status')['default'].mean().reset_index()
tab1['meaning'] = tab1['checking_status'].map(checking_map)
tab2 = eda_df.groupby('credit_history')['default'].mean().reset_index()
tab2['meaning'] = tab2['credit_history'].map(history_map)
tab3 = eda_df.groupby('savings_account')['default'].mean().reset_index()
tab3['meaning'] = tab3['savings_account'].map(savings_map)
tab4 = eda_df.groupby('employment_since')['default'].mean().reset_index()
tab4['meaning'] = tab4['employment_since'].map(employment_map)

display(tab1.sort_values('default', ascending=False))
display(tab2.sort_values('default', ascending=False))
display(tab3.sort_values('default', ascending=False))
display(tab4.sort_values('default', ascending=False))
```

**Interpretation:**
- Higher risk in `A11/A12` means low checking balances are linked to default.
- Higher risk in `A33/A34` means delayed/critical credit history is linked to default.
- Higher risk in `A61/A65` means low or unknown savings is linked to default.
- Higher risk in `A71/A72` means unemployment/short employment is linked to default.

## 5. BASELINE MODEL: LOGISTIC REGRESSION

A ColumnTransformer was implemented to apply targeted transformations to specific feature groups. Categorical variables were converted into binary dummy variables using OneHotEncoder, while numerical variables were normalized using StandardScaler() to ensure all features contribute equally to the model. These steps were encapsulated within a Pipeline to bundle preprocessing and the Logistic Regression model together.

The Logistic Regression model was configured with L2 regularization (penalty='l2') to prevent overfitting by penalizing large coefficients. To ensure the optimization algorithm reached a stable solution, max_iter was increased to 1000.

```python
preprocess_logit = ColumnTransformer([
    ('num', StandardScaler(), numeric_cols),
    ('cat', OneHotEncoder(handle_unknown='ignore'), categorical_cols)
])

logit_model = Pipeline([
    ('prep', preprocess_logit),
    ('clf', LogisticRegression(penalty='l2', max_iter=1000, random_state=RANDOM_STATE))
])

logit_model.fit(X_train, y_train)
logit_pred = logit_model.predict(X_test)
logit_prob = logit_model.predict_proba(X_test)[:, 1]

logit_accuracy = accuracy_score(y_test, logit_pred)
logit_precision = precision_score(y_test, logit_pred, zero_division=0)
logit_recall = recall_score(y_test, logit_pred, zero_division=0)
logit_f1 = f1_score(y_test, logit_pred, zero_division=0)
logit_auc = roc_auc_score(y_test, logit_prob)

print('Logit metrics')
print('Accuracy :', round(logit_accuracy, 3))
print('Precision:', round(logit_precision, 3))
print('Recall   :', round(logit_recall, 3))
print('F1-score :', round(logit_f1, 3))
print('ROC-AUC  :', round(logit_auc, 3))
```

```python
cm = confusion_matrix(y_test, logit_pred)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', cbar=False)
plt.title('Confusion Matrix - Logit')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()

fpr_logit, tpr_logit, _ = roc_curve(y_test, logit_prob)
plt.plot(fpr_logit, tpr_logit, label=f'Logit AUC={logit_auc:.3f}', linewidth=2)
plt.plot([0, 1], [0, 1], '--', color='gray')
plt.title('ROC - Logit')
plt.xlabel('FPR')
plt.ylabel('TPR')
plt.legend()
plt.show()
```

```python
feature_names = logit_model.named_steps['prep'].get_feature_names_out()
coefs = logit_model.named_steps['clf'].coef_[0]

coef_table = pd.DataFrame({'feature': feature_names, 'coef': coefs, 'odds_ratio': np.exp(coefs)})
coef_table['abs_coef'] = coef_table['coef'].abs()
coef_table = coef_table.sort_values('abs_coef', ascending=False)
display(coef_table[['feature', 'coef', 'odds_ratio']].head(20))
```

**Interpretation:**
- `cat__checking_status_A11` positive -> `<0 DM` checking balance increases default odds.
- `cat__credit_history_A33` positive -> past delays increase default odds.
- `cat__savings_account_A61` positive -> very low savings increases default odds.
- `cat__employment_since_A71` positive -> unemployment increases default odds.

## 6. MACHINE LEARNING MODELS

```python
preprocess_tree = ColumnTransformer([
    ('num', 'passthrough', numeric_cols),
    ('cat', OneHotEncoder(handle_unknown='ignore'), categorical_cols)
])

rf_model = Pipeline([
    ('prep', preprocess_tree),
    ('clf', RandomForestClassifier(n_estimators=300, class_weight='balanced', random_state=RANDOM_STATE))
])

rf_model.fit(X_train, y_train)
rf_pred = rf_model.predict(X_test)
rf_prob = rf_model.predict_proba(X_test)[:, 1]

rf_accuracy = accuracy_score(y_test, rf_pred)
rf_precision = precision_score(y_test, rf_pred, zero_division=0)
rf_recall = recall_score(y_test, rf_pred, zero_division=0)
rf_f1 = f1_score(y_test, rf_pred, zero_division=0)
rf_auc = roc_auc_score(y_test, rf_prob)

print('RF ROC-AUC:', round(rf_auc, 3))
```

```python
xgb_model = Pipeline([
    ('prep', preprocess_tree),
    ('clf', XGBClassifier(
        n_estimators=300,
        learning_rate=0.05,
        max_depth=4,
        subsample=0.9,
        colsample_bytree=0.9,
        objective='binary:logistic',
        eval_metric='logloss',
        random_state=RANDOM_STATE
    ))
])

xgb_model.fit(X_train, y_train)
xgb_pred = xgb_model.predict(X_test)
xgb_prob = xgb_model.predict_proba(X_test)[:, 1]

xgb_accuracy = accuracy_score(y_test, xgb_pred)
xgb_precision = precision_score(y_test, xgb_pred, zero_division=0)
xgb_recall = recall_score(y_test, xgb_pred, zero_division=0)
xgb_f1 = f1_score(y_test, xgb_pred, zero_division=0)
xgb_auc = roc_auc_score(y_test, xgb_prob)

print('XGB ROC-AUC:', round(xgb_auc, 3))
```

```python
tree_features = rf_model.named_steps['prep'].get_feature_names_out()

var_labels = {
    'checking_status': 'Checking account status',
    'credit_history': 'Credit history',
    'purpose': 'Purpose',
    'savings_account': 'Savings account',
    'employment_since': 'Employment since',
    'installment_rate': 'Installment rate (% disposable income)',
    'personal_status_sex': 'Personal status and sex',
    'other_debtors': 'Other debtors/guarantors',
    'residence_since': 'Present residence since',
    'property_type': 'Property',
    'age_years': 'Age (years)',
    'other_installment_plans': 'Other installment plans',
    'housing': 'Housing',
    'existing_credits_bank': 'Existing credits at this bank',
    'job_type': 'Job',
    'liable_people': 'People liable for maintenance',
    'telephone': 'Telephone',
    'foreign_worker': 'Foreign worker',
    'duration_months': 'Duration (months)',
    'credit_amount': 'Credit amount'
}

code_labels = {
    'A11': '< 0 DM (A11)',
    'A12': '0-200 DM (A12)',
    'A13': '>= 200 DM (A13)',
    'A14': 'No checking account (A14)',
    'A30': 'No credits/all paid duly (A30)',
    'A31': 'All credits at this bank paid duly (A31)',
    'A32': 'Existing credits paid duly till now (A32)',
    'A33': 'Delay in paying off in the past (A33)',
    'A34': 'Critical account/other credits existing (A34)',
    'A61': '< 100 DM (A61)',
    'A62': '100-500 DM (A62)',
    'A63': '500-1000 DM (A63)',
    'A64': '>= 1000 DM (A64)',
    'A65': 'Unknown/no savings account (A65)',
    'A71': 'Unemployed (A71)',
    'A72': '< 1 year (A72)',
    'A73': '1-4 years (A73)',
    'A74': '4-7 years (A74)',
    'A75': '>= 7 years (A75)',
    'A21': 'real state (A121)',    
    'A124': 'unknown/no property (A124)',
    'A151': 'Rent (A151)',
    'A152': 'Own (A152)',
    'A153': 'For free (A153)',
    'A201': 'Yes (A201)',
    'A202': 'No (A202)'
}

def readable_feature_name(feature):
    if feature.startswith('num__'):
        col = feature.replace('num__', '')
        return var_labels.get(col, col)
    if feature.startswith('cat__'):
        rest = feature.replace('cat__', '')
        var, code = rest.rsplit('_', 1)
        var_name = var_labels.get(var, var)
        code_name = code_labels.get(code, code)
        return f"{var_name}: {code_name}"
    return feature

rf_imp = pd.DataFrame({'feature': tree_features, 'importance': rf_model.named_steps['clf'].feature_importances_}).sort_values('importance', ascending=False)
xgb_imp = pd.DataFrame({'feature': tree_features, 'importance': xgb_model.named_steps['clf'].feature_importances_}).sort_values('importance', ascending=False)

rf_imp['feature_readable'] = rf_imp['feature'].apply(readable_feature_name)
xgb_imp['feature_readable'] = xgb_imp['feature'].apply(readable_feature_name)

fig, axes = plt.subplots(1, 2, figsize=(16, 6))
sns.barplot(data=rf_imp.head(10), x='importance', y='feature_readable', ax=axes[0], palette='crest')
axes[0].set_title('Random Forest - Top 10 important variables')
axes[0].set_ylabel('Variable')
sns.barplot(data=xgb_imp.head(10), x='importance', y='feature_readable', ax=axes[1], palette='mako')
axes[1].set_title('XGBoost - Top 10 important variables')
axes[1].set_ylabel('Variable')
plt.tight_layout()
plt.show()
```

**Interpretation:** if the most important variables include `A11/A12` (checking account < 0 DM; 0-200 DM), `A33/A34` (past payment delays; critical account/other existing credits), and `A61/A65` (savings < 100 DM; unknown savings/no savings account), then the models are detecting profiles with lower liquidity and poorer credit history as the primary drivers of default risk. If `A71/A72` (unemployed; employment < 1 year) also appear, it suggests that labor fragility is associated with higher risk.

## 7. MODEL COMPARISON

```python
comparison = pd.DataFrame({
    'Model': ['Logistic Regression', 'Random Forest', 'XGBoost'],
    'Accuracy': [logit_accuracy, rf_accuracy, xgb_accuracy],
    'Precision': [logit_precision, rf_precision, xgb_precision],
    'Recall': [logit_recall, rf_recall, xgb_recall],
    'F1': [logit_f1, rf_f1, xgb_f1],
    'ROC_AUC': [logit_auc, rf_auc, xgb_auc]
}).sort_values('ROC_AUC', ascending=False)
display(comparison)
```

```python
fpr_rf, tpr_rf, _ = roc_curve(y_test, rf_prob)
fpr_xgb, tpr_xgb, _ = roc_curve(y_test, xgb_prob)

plt.figure(figsize=(7, 6))
plt.plot(fpr_logit, tpr_logit, label=f'Logit AUC={logit_auc:.3f}', linewidth=2)
plt.plot(fpr_rf, tpr_rf, label=f'RF AUC={rf_auc:.3f}', linewidth=2)
plt.plot(fpr_xgb, tpr_xgb, label=f'XGB AUC={xgb_auc:.3f}', linewidth=2)
plt.plot([0, 1], [0, 1], '--', color='gray')
plt.title('ROC comparison')
plt.xlabel('FPR')
plt.ylabel('TPR')
plt.legend()
plt.show()
```

If a more complex Machine Learning model increases the ROC-AUC or F1-Score, it means the model has become better at distinguishing between "Good" and "Bad" credit risks, especially by balancing Precision and Recall. However, while these models might catch more defaults, they often act as a "black box," making it difficult to pinpoint exactly why a specific loan was denied.

In contrast, because Logistic Regression (Logit) produces direct coefficients and Odds Ratios, you can provide a clear explanation.

## 8. CONCLUSIONS

The selection of the best model is determined by the highest out-of-sample ROC-AUC and F1-Score, which collectively measure the model's ability to distinguish between "Good" and "Bad" risks while balancing precision and recall.

- Logistic Regression: best interpretability.
- RF/XGB: higher predictive accuracy by capturing complex, nonlinear patterns and feature interactions that a linear model might miss.

The riskier profile is:
- `A11/A12` (<0 DM/0-200 DM) in checking account, `A61/A65` (<100 DM/no savings account) in savings.
- `A33/A34` (delay in paying off in the past/ critical account) in credit history.
- `A71/A72` (unemployed/<1 year) in employment.
